```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot of service levels by service type
sns.countplot(x='Service Type', hue='Service level', data=df)
plt.title('Service Levels by Service Type')
plt.xlabel('Service Type')
plt.ylabel('Count')
plt.show()

# Create a line plot of coverage by year
sns.lineplot(x='Year', y='Coverage', data=df)
plt.title('Coverage by Year')
plt.xlabel('Year')
plt.ylabel('Coverage')
plt.show()

# Create a scatter plot of coverage by school-age population
sns.scatterplot(x='School-age population', y='Coverage', data=df)
plt.title('Coverage by School-age Population')
plt.xlabel('School-age Population')
plt.ylabel('Coverage')
plt.show()
```
